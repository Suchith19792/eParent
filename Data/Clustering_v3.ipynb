```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = [15,10]
```

```python
df = pd.read_csv("Dummydata_v2.csv",index_col='id')
```

```python
df.head()
```

```python
df.describe()
```

```python
math = df.loc[:,"grade1_maths":"grade10_maths"]
science = df.loc[:,"grade1_science":"grade10_science"]
language = df.loc[:,"grade1_language":"grade10_language"]
```

```python
df['avg_math'] = math.mean(axis=1)
df['avg_science'] = science.mean(axis=1)
df['avg_language'] = language.mean(axis=1)
```

```python
df.describe()
```

```python
dfs = df.loc[:,"avg_math":"avg_language"]
dfs.describe()
```

```python
df4c = df.loc[:,"grade1_maths":"grade10_language"]
df4c.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('Dendrograms (ward)')
dend = shc.dendrogram(shc.linkage(df4c,method='ward'))
plt.axhline(y=1600,color='r',linestyle='--')
plt.savefig('dendrogram_ward.png')
```

```python
from sklearn.cluster import AgglomerativeClustering
clust_hac_ward = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hac_ward = clust_hac_ward.fit_predict(df4c)
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrograms (average)')
dend = shc.dendrogram(shc.linkage(df4c,method='average'))
plt.axhline(y=180,color='r',linestyle='--')
plt.savefig('dendrogram_average.png')
```

```python
clust_hac_avg = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='average')
hac_avg = clust_hac_avg.fit_predict(df4c)
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrograms (complete)')
dend = shc.dendrogram(shc.linkage(df4c,method='complete'))
plt.axhline(y=400,color='r',linestyle='--')
plt.savefig('dendrogram_complete.png')
```

```python
clust_hac_cmplt = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hac_cmplt = clust_hac_cmplt.fit_predict(df4c)
```

```python
df['label_ward'] = clust_hac_ward.labels_
df['label_avg'] = clust_hac_avg.labels_
df['label_cmplt'] = clust_hac_cmplt.labels_
```

```python
df.loc[:,'label':'label_cmplt'].head(10)
```

```python
df.groupby('label').mean()
```

```python
dft = df.groupby('label').mean().transpose()
dft.head()
```

```python
dft.iloc[0:10].plot.bar(rot=0);
plt.savefig('ClusterPerf_math.png')
```

```python
dft.iloc[10:20].plot.bar(rot=0);
plt.savefig('ClusterPerf_science.png')
```

```python
dft.iloc[20:30].plot.bar(rot=0);
plt.savefig('ClusterPerf_language.png')
```
