```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Dummydata_v2.csv",index_col='id')
```

```python
df.head()
```

```python
df.describe()
```

```python
math = df.loc[:,"grade1_maths":"grade10_maths"]
science = df.loc[:,"grade1_science":"grade10_science"]
language = df.loc[:,"grade1_language":"grade10_language"]
```

```python
df['avg_math'] = math.mean(axis=1)
df['avg_science'] = science.mean(axis=1)
df['avg_language'] = language.mean(axis=1)
```

```python
df.describe()
```

```python
dfs = df.loc[:,"avg_math":"avg_language"]
dfs.describe()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('Dendrograms')
dend = shc.dendrogram(shc.linkage(dfs,method='ward'))
plt.axhline(y=500,color='r',linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
cluster.fit_predict(dfs)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dfs['avg_math'],dfs['avg_science'],c=cluster.labels_)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dfs['avg_science'],dfs['avg_language'],c=cluster.labels_)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dfs['avg_language'],dfs['avg_math'],c=cluster.labels_)
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrograms')
dend2 = shc.dendrogram(shc.linkage(dfs,method='complete'))
plt.axhline(y=90,color='r',linestyle='--')
```
